```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import datetime
```

```python
URL = "https://www.amazon.in/s?k=iphone&crid=3UKKNKQT5XYKI&sprefix=iphon%2Caps%2C273&ref=nb_sb_ss_pltr-xclick_2_5"
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36 Edg/126.0.0.0"}
```

```python
webpage = requests.get(URL,headers= headers)
webpage
```

```python
soup = BeautifulSoup(webpage.content,"html.parser")
```

```python
links = soup.find_all('a',attrs={"class":"a-link-normal s-no-outline"})
```

```python
def get_title(soup2):
    try:
        title = soup2.find("span",attrs={"id":"productTitle"}).text.strip()
        return title
    except Exception:
        return ""

def get_price(soup2):
    try:
        price = soup2.find("span",attrs={"class":"a-price-whole"}).text.strip()
        return price
    except Exception:
        return ""

def get_ratings(soup2):
    try:
        ratings = soup2.find("span",attrs={"class":"a-icon-alt"}).text.strip()
        return ratings
    except Exception:
        return ""

def get_reviews(soup2):
    try:
        reviews = soup2.find("span",attrs={"id":"acrCustomerReviewText"}).text.strip()
        return reviews
    except Exception:
        return ""

def get_availability(soup2):
    try:
        availability = soup2.find("span",attrs={"class":"a-size-medium a-color-success"}).text.strip()
        return availability
    except Exception:
        return ""
```

```python
import pandas as pd

product_link = []
for link in links:
    href = link.get("href")
    if href.startswith("/"):
        product_link.append("https://www.amazon.in" + href)
    else:
        continue

details = {"title":[], "price":[], "ratings":[], "reviews":[], "availability":[]}

for link in product_link:
    new_webpage = requests.get(link,headers=headers)
    soup2 = BeautifulSoup(new_webpage.content, "html.parser")

    details['title'].append(get_title(soup2))
    details['price'].append(get_price(soup2))
    details['ratings'].append(get_ratings(soup2))
    details['reviews'].append(get_reviews(soup2))
    details['availability'].append(get_availability(soup2))

df = pd.DataFrame(details)
Amazon_csv = df.to_csv("C:/Users/jkrat/Downloads/amazon_iphone_data.csv", header = True, index = False)
df
```
